```python
import pickle
import pandas as pd
import re 
import string
text_files = ["a.txt", "b.txt", "c.txt", "d.txt"]

speaker1 = []
speaker2 = []
speaker3 = []
speaker4 = []
speaker5 = []
speaker6 = []
speaker7 = []
speaker8 = []
listoflists = [speaker1, speaker2, speaker3, speaker4, speaker5, speaker6, speaker7, speaker8]
da = {}
# opens the text file with with lines denoted as "\n"        
for i in text_files:
    speaker1 = []
    speaker2 = []
    
    with open(i) as f:
        list0 = f.readlines()
        
        
        for lines in list0:
            
            # get rid of "\n" and punctuation
            if lines[:10] == "speaker 1:":
                lines = re.sub('\n', '', lines)
                lines = re.sub('\[.*?\]', '', lines)
                speaker1.append( re.sub('[%s]'%re.escape(string.punctuation), '', lines[10:].lower()))
            if lines[:10] == "speaker 2:":
                lines = re.sub('\n', '', lines)
                lines = re.sub('\[.*?\]', '', lines)
                speaker2.append( re.sub('[%s]'%re.escape(string.punctuation), '', lines[10:].lower()))
                
    da[i] = [speaker1, speaker2]

df = pd.DataFrame(da)
df
```

### data



```python
# create a corpus (collection of text document-term matrix): word counts in matrix format
file_names = [ 'a', 'b', 'c', 'd' ]
#!mkdir transcripts
transcript = []
for i in df:
    transcript.append(df[i][0])
    transcript.append(df[i][1])
    
```

```python
text_files = ["speaker1 _a.txt", "speaker2_a.txt", "speaker1_b.txt", "speaker2_b.txt", "speaker1_c.txt", "speaker2_c.txt", "speaker1_d.txt", "speaker2_d.txt" ]

for i, c in enumerate(text_files):
    with open("transcripts/"+c+".txt", "wb") as file:
        pickle.dump(transcript[i], file)
```

```python
# load pickled files
data = {}
for i, c in enumerate(text_files):
    with open("transcripts/"+c+".txt", "rb") as file:
        data[c] = pickle.load(file)
        
# we are going to create a document-term matrix using CountVectorizer, 
# exclude common english stop words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# we will change this to | key: speaker, value: string | format
def combined_text(list_of_text):
    # takes a list of txt and combine them into a large chunck of text
    combined_text = " ".join(list_of_text)
    return combined_text

# combine the data
data_combined = {key: [combined_text(value)] for (key, value) in data.items()}
#data_combined

# throw it into a pandas data frame
import pandas as pd
pd.set_option("max_colwidth", 150)

data_df = pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns = ["transcripts"]
data_df = data_df.sort_index()
data_df
```

```python
cv = CountVectorizer(stop_words="english")
data_cv = cv.fit_transform(data_df.transcripts)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_df.index
data_dtm
```

```python
# pickle for later use
data_dtm.to_pickle("dtm.pkl")
```

```python
# pickle the cleaned data and the CountVectorizer object
data_df.to_pickle("data_df.pkl")
pickle.dump(cv, open("cv.pkl", "wb"))
```
